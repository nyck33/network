```c
// file_copy.c
/* copy file to another file*/

#include <stdio.h>
#include <stdlib.h>
//need 4 headers for IO functions
#include <sys/types.h>
#include <sys/stat.h>
#include <fcntl.h>
#include <unistd.h>

/* argv[1]: from, argv[2] to */

int main(int argc, char* argv[]){
    int in, out;
    
    in = open(argv[1], O_RDONLY);  //O_RDONLY is defined in /usr/include/sys/types.h
    if(in < 0){
        printf("Cannot open this file %s\n", argv[1]);
        exit(1);
    }
    out = open(argv[2], O_WRONLY | O_TRUNC | O_CREAT, S_IRUSR | S_IWUSR);
    if (out < 0){
        printf("Cannot create the file %s\n", argv[2]);
        exit(1);
    }
    //read bytes from in and write to out
    //print those bytes on to screen while copying
    
    close(in);
    close(out);
    //test to match content, print result
}
```

```c
#include <stdio.h>
#include <stdlib.h>

// copy string from source to dest
void string_copy(char *destination, char *source){
    char *s = NULL:
    s = source;
    #allocate memory of size source for destination
    count = 0;
    while(*s){
        count++;
        s++;
    }
    char dest[count];
    int counter=0;
    while(*source){
        dest[counter] = *source;
        source++;
    }
    destination = &dest;
}

//count string length
int string_length (char *s){
    int count = 0;
    while(*s!='\0'){
        s++;
        count++;
    }
    return count;
}

//concatenate source to end of dest
void string_cat (char *destination, char *source){
    int count=0;
    //initialized temp ptrs to point to dest and source
    char *d = NULL;
    d = destination;
    char *s = NULL;
    s = source;
    //advance pointer to end of destination on top of endline
    //pointer to endline and breaks 
    while(*d!='\0'){
        count++;
        d++;
    }
    //while source not empty
    while(*s){
        count++;
        s++;
    }
    //initialize char array
    char total[count];
    //new counter
    counter = 0;
    //copy to new array
    while(*destination!='\0'){
        total[counter]=*destination;
        destination++;
        counter++;
    }
    while(*source){
        total[counter]=*source;
        source++;
        counter++;
    }
    //set ptr dest to start of char total[]
    destination = &total;
}

```
