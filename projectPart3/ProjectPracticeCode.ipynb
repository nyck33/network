## uint to char to uchar to uint?

```c
#include <stdio.h>
 
int main(void) {
    
	unsigned int S = 0;
	unsigned int R = 1;
	char S_char = S + '0';
	char R_char = R + '0';
 
	unsigned int S_conv = (unsigned int)(unsigned char)(S_char - '0');
	unsigned int R_conv = (unsigned int)(unsigned char)(R_char - '0');
 
 
	printf("%u %u %u %u", S_conv, R_conv, S_char, R_char);
	return 0;
}
```

## for ACK, seqnum int -> char -> uchar (checksum) -> int

```c
#include <stdio.h>
 
int main(void) {
	
	int S = 0;
	int R = 1;
    //put in char array
	char S_char = (char)S; 
	char R_char = (char)R;
    
    //take out for csum calc
	unsigned char S_uc = (unsigned char)S_char;
    unsigned char R_uc = (unsigned char)R_char;
    
    //compare with seqnum
    int S_int = (int)S_uc;
    int R_int = (int)R_uc;
 
 
	printf("%d %d", S_int, R_int );
	return 0;
}
```
